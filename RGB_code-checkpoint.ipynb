```python
from PIL import Image

# open the image
img = Image.open("1_1_1.jpg")

# get RGB values of each pixel
pixels = img.load()
width, height = img.size

# calculate average RGB values for the whole image
total_r = 0
total_g = 0
total_b = 0
num_pixels = width * height

for x in range(width):
    for y in range(height):
        r, g, b = pixels[x, y]
        total_r += r
        total_g += g
        total_b += b

avg_r = total_r // num_pixels
avg_g = total_g // num_pixels
avg_b = total_b // num_pixels

print(f"Average RGB values: R={avg_r}, G={avg_g}, B={avg_b}")
print((avg_r+avg_g+avg_b)/3)
```
